```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_csv("alcohol-consumption.csv")
```

```python
df
```

```python
top_countries = df.groupby("country")["total_consumption"].sum().sort_values(ascending=False).head(10)
```

```python
print(top_countries)
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=top_countries.index, y=top_countries.values, palette="viridis")
plt.title("Top 10 Countries by Total Alcohol Consumption",fontsize=35)
plt.xlabel("Country", fontsize = 12)
plt.ylabel("Total Consumption (Litres per capita)", fontsize = 12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_unrecorded = df.groupby("country")["unrecorded_consumption"].sum().sort_values(ascending=False).head(10)
print(top_unrecorded)
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=top_unrecorded.index, y=top_unrecorded.values, palette = "viridis")
plt.title("Top 10 Countries by Unrecorded Alcohol Consumption", fontsize = 35)
plt.xlabel("Countries",fontsize = 12)
plt.ylabel("Unrecorded Alcohol Consumption (Litres per capita)", fontsize =12)
plt.xticks(rotation = 45)
plt.show()
```

```python
alcohol_types = ["beer_percentage","wine_percentage","spirits_percentage","other_percentage"]
```

```python

df_plot = df.set_index("country")[alcohol_types + ["total_consumption"]]
df_plot = df_plot.sort_values("total_consumption", ascending=False).head(10)
df_plot_percent = df_plot[alcohol_types]
plt.figure(figsize=(12,6))
df_plot_percent.plot(kind="bar",stacked=True,figsize=(12,6),colormap="tab20")
plt.title("Alcohol Composition of Top 10 Countries by Total_consumption")
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.legend(title="Alcohol Type")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sb.scatterplot(data=df, x="2020_projection", y="2025_projection",hue="country")
plt.title("2020 v 2025 Projected Alcohol Consumption")
plt.xlabel("2020 Projection")
plt.ylabel("2025 Projection")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
